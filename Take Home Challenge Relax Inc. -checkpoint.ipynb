```python
# Importing Packages

import pandas as pd
import numpy as np
```

```python
# Creating pandas dataframe with user related data

users = pd.read_csv('takehome_users.csv', encoding='latin-1', index_col=0)
users.head(10)
```

```python
# Creating pandas dataframe with user engagement data 

engagement = pd.read_csv('takehome_user_engagement.csv')
engagement.head(10)
```

```python
# Checking for datatypes & missing information
# last_session_creation_time has missing values which makes sense as not all users will have a last session
# invited_by_user_id has missing values whcih males sense as well as not all users are invited by other users

users.info()
```

```python
# Converting creation_time from string to datetime type

users['creation_time'] = pd.to_datetime(users['creation_time'])
```

```python
engagement.info()
```

```python
# Converting time_stamp from string to datetime type

engagement['time_stamp'] = pd.to_datetime(engagement['time_stamp'])
```

```python
# Setting time_stamp as index for engagement dataframe

engagement = engagement.set_index('time_stamp')
engagement.head(10)
```

```python
# Grouping data by user_id, resampling into weekly dates, & calculating sum in visited column to capture 'adopted' users

engagement = engagement.groupby('user_id').resample('1W').sum()
engagement
```

```python
adopted = engagement[engagement['visited'] >= 3]
del adopted['user_id']
adopted
```

```python
# Creating user_id list of adopted users

adopted_index = list(adopted.index)
adopted_users_list = []
for i in range(len(adopted_index)):
    adopted_users_list.append(adopted_index[i][0])
len(adopted_users_list)
```

```python
# Keeping set of only unique user ids

adopted_users = set(adopted_users_list)
len(adopted_users)
```

```python
# Adding adopted column to users dataframe

users['adopted'] = False

for user in adopted_users:
    users.loc[user, 'adopted'] = True

users.head(10)
```

```python
# Extracting domain names from email addresses

users['email_split'] = users['email'].str.split('@')
del users['email']

for user in range(len(users)):
    users.loc[user+1, 'email_domain'] = users.loc[user+1, 'email_split'][1]
del users['email_split']


users['email_domain'].value_counts()[:10]
```

```python
# Combining uncommon email domains into 'other' value

users['email'] = 'other'

for user in range(len(users)):
    if users.loc[user+1, 'email_domain'] == 'gmail.com':
        users.loc[user+1, 'email'] = 'gmail.com'
    if users.loc[user+1, 'email_domain'] == 'yahoo.com':
        users.loc[user+1, 'email'] = 'yahoo.com'
    if users.loc[user+1, 'email_domain'] == 'jourrapide.com':
        users.loc[user+1, 'email'] = 'jourrapide.com'
    if users.loc[user+1, 'email_domain'] == 'cuvox.de':
        users.loc[user+1, 'email'] = 'cuvox.de'
    if users.loc[user+1, 'email_domain'] == 'gustr.com':
        users.loc[user+1, 'email'] = 'gustr.com'
    if users.loc[user+1, 'email_domain'] == 'hotmail.com':
        users.loc[user+1, 'email'] = 'hotmail.com'

del users['email_domain']

users['email'].value_counts()
```

```python
# Creating invited_by_user column with boolean values based on whether or not a user was invited by another user

users['invited_by_user'] = users['invited_by_user_id'].isnull() == False
del users['invited_by_user_id']
users.head(10)
```

```python
# Converting users dataframe into suitable form for machine learning models 

users['org_id'] = users['org_id'].astype('str')
dummies = pd.get_dummies(users[['creation_source', 'org_id', 'email']])
users = pd.concat([users, dummies], axis=1)
```

```python
# Removing creation_time & last_session_creation_time as I believe these time variables are not needed/will make models unneccesarily complicated

users = users.drop(['name','creation_time', 'last_session_creation_time', 'creation_source', 'org_id', 'email'], axis=1)
```

```python
users.head(5)
```

```python
# Importing packages for machine learning models

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import GridSearchCV
```

```python
# Separating predictor variables from 'adopted' variable 

X = users.drop(['adopted'], axis=1)
Y = users.adopted
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=6)
```

```python
# Running standard random forest classifier model 

rfc = RandomForestClassifier(n_estimators=10, class_weight = 'balanced_subsample')
rfc.fit(X_train, Y_train)
rfc.score(X_train, Y_train)
```

```python
rfc.score(X_test, Y_test)
```

```python
# See that only 27 out of 420 adopted users identified correctly by standard classifier

Y_pred = rfc.predict(X_test)
print(classification_report(Y_test, Y_pred))
```

```python
print(confusion_matrix(Y_test, Y_pred))
```

```python
# Feature importances within standard classifer 

feature_importances = pd.DataFrame({'importances': rfc.feature_importances_}, index=X_train.columns)
feature_importances.sort_values('importances', ascending=False).head(10)
```

```python
# Tuning random forest classifier model

params = {'n_estimators': [10], 'max_depth': [2, 3, 4, 5], 'class_weight': ['balanced_subsample']}

rfc_tuned = RandomForestClassifier()
rfc_grid = GridSearchCV(rfc_tuned, params, cv=3, scoring='accuracy')
rfc_grid.fit(X_train, Y_train)
rfc_grid.score(X_train, Y_train)
```

```python
rfc_grid.score(X_test, Y_test)
```

```python
rfc_grid.best_estimator_
```

```python
# Random forest classifer model with best estimators

rfc_final = RandomForestClassifier(n_estimators=10, max_features='auto', min_samples_leaf=1, min_samples_split=2, class_weight='balanced_subsample', max_depth=4)
rfc_final.fit(X_train, Y_train)
rfc_final.score(X_train, Y_train)
```

```python
rfc_final.score(X_test,Y_test)
```

```python
# 224 out of 420 adopted users identified correcly 
# More tuning needs to be done to increase correct identification of adopted users 

Y_pred = rfc_final.predict(X_test)
print(classification_report(Y_test, Y_pred))
```

```python
print(confusion_matrix(Y_test, Y_pred))
```

```python
# Feature importances in tuned model 

feature_importances_final = pd.DataFrame({'importances': rfc_final.feature_importances_}, index=X_train.columns)
feature_importances_final.sort_values('importances', ascending=False).head(20)
```

```python
# Feature imporance differences will show which attributes are responsible for increasing detection of adopted feature_importance_diff = pd.DataFrame(index=X_train.columns)

feature_importance_diff['standard_importance'] = rfc.feature_importances_
feature_importance_diff['final_importance'] = rfc_final.feature_importances_
feature_importance_diff['difference'] = feature_importance_diff['final_importance'] - feature_importance_diff['standard_importance']
feature_importance_diff.sort_values('difference', ascending=False).head(20)
```
